¿Cuántas personas lo usan cada día, semana y mes?

```python
# Cálculo de usuarios diarios activos
dau_total = df_visits.groupby('visits_date').agg({'uid': 'nunique'})

# Cálculo de usuarios semanales activos
wau_total = df_visits.groupby(['visits_week']).agg({'uid': 'nunique'})

# Cálculo de usuarios mensuales activos
mau_total = df_visits.groupby('visits_month').agg({'uid': 'nunique'})
```

```python
# Creación de gráfico de una linea
plt.figure(figsize=(10, 5))
plt.plot(dau_total.index, dau_total.values, label='Usuarios diarios', alpha=0.15)

# Promedio rolling semanal 
plt.plot(dau_total.index, dau_total.rolling(window=7).mean(),
         label='Weekly Rolling Average', alpha=0.7)

# Promedio rolling mensual
plt.plot(dau_total.index, dau_total.rolling(window=30).mean(),
         label='Monthly Rolling Average', alpha=0.7)


plt.xlabel('Date')
plt.ylabel('Unique Users')
plt.title('Daily and Weekly Rolling Average of Unique Users')
plt.legend()
plt.show()
```


En esta gráfica se muestra la duración promedio de las sesiones semanal y mensual, así como los usuarios diarios, este gráfico de líneas permite observar con más detalle la duración promedio según el periodo de tiempo, se pueden observar algunos picos de aumento del promedio lo cual indica que en ese periodo de tiempo el departamento de marketing implementó algunas estrategia que lograron aumentar el tiempo promedio que permanece el usuario, habría que revisar las estrategias implementadas y con esto demostrar que están funcionando.

¿Cuándo la gente empieza a comprar?
(En el análisis de KPI, generalmente nos interesa saber el tiempo que transcurre entre el registro y la conversión, es decir, cuando el/la usuario/a se convierte en cliente. Por ejemplo, si el registro y la primera compra ocurren el mismo día, el/la usuario/a podría caer en la categoría Conversion 0d. Si la primera compra ocurre al día siguiente, será Conversion 1d.
```python
# Cálculo para obtener primera fecha de compra de los clientes
orders = df_orders.assign(dt_1st_purchase=df_orders.groupby('uid')['buy_ts'].transform('min'))

# Tiempo transcurrido hasta que se convierten en clientes
orders['time_passed'] = orders['buy_ts'] - orders['dt_1st_purchase']
orders['time_passed'] /= np.timedelta64(1, 'h')  # Se convierte en horas

# Se crea la columna first order month
orders['first_order_month'] = orders['dt_1st_purchase'].dt.to_period('M').dt.start_time

print(orders.head())
```

```python
# Se hace un merge para obtener las columnas necesarias para los calculos
romi_time = pd.merge(orders, df_visits, on='uid', how='left').drop(columns=['device', 'start_ts', 'visits_date', 
                                                                            'visits_week', 'visits_month', 
                                                                            'session_duration', 'time_between_visits', 'first_month',
                                                                            'first_day', 'visit_age', 'end_ts'])

# Se agrupan los datos por primes mes de compra y mes de otras compras
cohorts = (romi_time.groupby(['first_order_month', 'month']).agg({'revenue': 'sum'}).reset_index())

# Se agrupan los datos para obtener el tamaño de los cohortes
cohort_sizes = (romi_time.groupby(['first_order_month', 'source_id']).agg({'uid': 'nunique'}).reset_index())

# Se hace otro merge para poder obtener las columnas nececsarias para calculo de ltv
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')

# Se crea la columna ltv_cohort para obtener el ltv por cada 
report['ltv_cohort'] = report['revenue'] / report['uid']

# Se hace otro merge para obtener la columna de costos
table_romi = pd.merge(report, df_cost, on='source_id', how='left').drop(columns=['date', 'day', 'week', 'cohorte'])

# Se crea la columna para calular el cac por cohort 
table_romi['cac_cohort'] = (table_romi['costs'] / table_romi['uid'])

# Se crea la columna para calular el romi por cohort 
table_romi['romi_cohort'] = (table_romi['ltv_cohort'] / table_romi['cac_cohort'])
table_romi['first_order_month'] = table_romi['first_order_month'].dt.to_period('M')
```

```python
# Se crea la pivot_table para mostrar el ROMI por cohorte
table_romi_pivot = table_romi.pivot_table(
    index='first_order_month',
    columns='source_id',
    values='romi_cohort',
    aggfunc='sum')

# Se crea un heatmap para mostrar la informacion
plt.figure(figsize=(13, 9))
sns.heatmap(table_romi_pivot, annot=True, fmt='.2f',
    linewidths=1,
    linecolor='black', cmap=cmap)
plt.title('ROMI por cohorte')
plt.show()
```

  - Con eso calculamos un revenue por comprador de cada mes para cada cohorte
  - Además, obtén los costos mensuales de la source i y únelos como mes de primera orden = mes de gasto
  - Así, tendremos ltv=revenue/tamaño cohorte y cac=gasto/tamaño cohorte
  - Finalmente ROMI = ltv/cac

```python
# Primera visita usuarios
first_visit_users = (
    df_visits[['uid','start_ts','source_id']]
    .sort_values(by='start_ts',ascending=True)
    .groupby('uid',as_index=False)
    .first()
    )

# Mes de primera visita usuarios
first_visit_users['first_visit_month'] = first_visit_users['start_ts'].values.astype('datetime64[M]')
    
# Primera orden usuarios
first_order_users = (
    df_orders[['uid','buy_ts']]
    .sort_values(by='buy_ts',ascending=True)
    .groupby('uid',as_index=False)
    .first()
    )
    
# Mes de primera orden usuarios
first_order_users['first_order_month'] = first_order_users['buy_ts'].values.astype('datetime64[M]')

# Seleccionar columnas a usar de visitas
first_visit_users = first_visit_users[['uid','source_id','first_visit_month']]

## Seleccionar columnas a usar de orders
first_order_users = first_order_users[['first_order_month','uid']]
    
# Unir data de primera orden a dataframe orders
df_orders = pd.merge(df_orders, first_order_users, on='uid', how='inner')
    
# Obtener mes de compra
df_orders['order_month'] = df_orders['buy_ts'].values.astype('datetime64[M]')
    
# Seleccionar columnas a usar de ordenes
buyers = df_orders[['uid','revenue','order_month']]

# Unir data de ordenes con data de visitas
buyers = pd.merge(buyers, first_visit_users, on='uid')

# Generar columna de mes en los costos
df_cost['costs_month'] = df_cost['date'].values.astype('datetime64[M]')
```

```python
# Se crea el ciclo for donde se hara el recorrido a traves de las 7 fuentes con la tabla buyers creada anteriormente
for source, df in buyers.groupby('source_id'):
    if source !=7:

# Se crea una agrupacion para filtrar los usuarios unicos que llegaron por cada fuente
        cohort_sizes_t = (
            df.groupby(['source_id', 'first_visit_month'])
            .agg({'uid': 'nunique'})
            .reset_index()
        )

        # Se seleccionan las columnas que se van a usar
        cohort_sizes_t.columns = ['source_id', 'first_order_month', 'n_buyers']

        # Se hace una agrupacion para obtener la sumatoria del revenue, por fuente, primera visita y mes de order
        cohorts_revenue_t = (
            df.groupby(['source_id', 'first_visit_month', 'order_month'])
            .agg({'revenue': 'sum'})
            .reset_index()
        )

        #Se seleccionan las columnas a usar
        cohorts_revenue_t.columns = ['source_id', 'first_order_month', 'order_month', 'gp']

        # Se hace un merge de las agrupaciones anteriores
        report_romi = pd.merge(cohort_sizes_t, cohorts_revenue_t, on=['source_id', 'first_order_month'])

        # Se hace el calculo de la columan age restando el mes de orden del mes de primera orden
        report_romi['age'] = np.round(((report_romi['order_month'] - report_romi['first_order_month']) / np.timedelta64(1, 'M')).astype('int'))
        report_romi['age'] = report_romi['age'].values.astype('int')

        #Tendremos ltv=revenue/tamaño cohorte
        report_romi['ltv'] = report_romi['gp'] / report_romi['n_buyers']
        
        # Aqui se hace un merge con el dataframe original de costos
        report_romi = pd.merge(report_romi, df_cost.groupby(['source_id', 'costs_month']).agg({'costs': 'sum'}).reset_index())
        report_romi = report_romi.drop(['costs_month'], axis = 1) 

        # cac=gasto/tamaño cohorte
        report_romi['cac'] = report_romi['costs'] / report_romi['n_buyers']

        #Finalmente ROMI = ltv/cac
        report_romi['romi'] = report_romi['ltv'] / report_romi['cac']

        # Se crea una tabla pivote de source 
        romi = report_romi\
                .pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean')\
                .(axis=1)
        romi.round(2).fillna('')

        romi.index=romi.index.astype(str)

        # Se grafica cada uno de los source
        sns.heatmap(romi, annot=True, fmt='.2f', linewidths=1, linecolor='grey', annot_kws={'size': 6}, cmap=cmap, cbar_kws= {'orientation': 'horizontal'} 
            ).set(title ='ROMI by source {}'.format(source))
        plt.show()
```
